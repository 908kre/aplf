```python
from ipywidgets import interact
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.palettes import Category10
output_notebook()
```

```python
import pandas as pd
from toolz.curried import pipe, map, concat
import os
fig_opt = {'x_axis_type': 'datetime'}
```

```python
dir = '/store/kaggle/nab/artificialNoAnomaly/artificialNoAnomaly/'
fns = pipe(os.walk(dir), map(lambda x: x[2]), concat, list)
p = figure(**fig_opt)
for fn, c in zip(fns, Category10[len(fns)]):
    full_path = f'{dir}/{fn}'
    df = pd.read_csv(full_path)
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    p.circle(df['timestamp'], df['value'], color=c, legend=fn)
show(p)
```
